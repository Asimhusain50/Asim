```python
df.drop(columns=['EmployeeCount','Education'],inplace = True)
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
```

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# what is pd.cut
df['Experience Level'] = pd.cut(
    df['TotalWorkingYears'],
    bins = [0,3,7,15,40],
    labels = ['Entry','Mid','Senior','Expert'])
```

```python
df['TenureGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins = [0,2,5,10,20,40],
    labels = ['0-2','3-5','6-10','11-20','20+'])
```

```python
df
```

```python
Attrition_trend  = df.groupby('TenureGroup')['Attrition_Flag'].mean().reset_index()
Attrition_trend['Attrition_rate (%)'] = Attrition_trend['Attrition_Flag'] * 100
Attrition_trend.drop(columns = ['Attrition_Flag'],inplace = True)

```

```python
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 4))
sns.barplot(x=attrition_rate.index, y=attrition_rate.values, palette='Set2')
plt.title("Overall Attrition Rate")
plt.ylabel("Percentage")
plt.xlabel("Attrition")
plt.show()
```

```python

sns.countplot(data=df, x='Department', hue='Attrition', palette='coolwarm')
```

```python
sns.boxplot(data = df , x = "Attrition" , y = "TenureGroup" )
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Attrition', y='JobSatisfaction', palette='Set1')
plt.title("Job Satisfaction by Attrition")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data = df , x = 'EducationField' , hue = 'Attrition' , palette = 'coolwarm')
plt.title('Attrition Rate according Education field')
plt.show()
```

```python
df['Attrition'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, 
                                        colors=['skyblue', 'salmon'],
                                        labels=['No', 'Yes'])

plt.title("Attrition Distribution")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
df['EducationField'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90,
                                             figsize=(12, 10), colors=plt.cm.Pastel1.colors)

plt.title("Distribution of Employees by Education Field")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
df['Department'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90,
                                             figsize=(12, 10), colors=plt.cm.Pastel1.colors)

plt.title("Distribution of Employees by Department")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
job_attrition = df.groupby('JobRole')['Attrition_Flag'].mean().sort_values(ascending=False) * 100

sns.barplot(x=job_attrition.values, y=job_attrition.index, palette='coolwarm')
plt.title("Attrition Rate by Job Role")
plt.xlabel("Attrition Rate (%)")
plt.ylabel("Job Role")
plt.show()
```

```python
# 1. Pie Chart: Overall Attrition
# -----------------------------
plt.figure(figsize=(6, 6))
df['Attrition'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90,
                                        labels=['Stayed', 'Left'],
                                        colors=['skyblue', 'salmon'])
plt.title("Overall Employee Attrition")
plt.ylabel("")
plt.show()

# -----------------------------
# 2. Bar Chart: Attrition by Department
# -----------------------------
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Department', hue='Attrition', palette='Set2')
plt.title("Attrition by Department")
plt.xticks(rotation=15)
plt.show()

# -----------------------------
# 3. KDE Plot: Monthly Income vs Attrition
# -----------------------------
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='MonthlyIncome', hue='Attrition', fill=True, palette='Set1')
plt.title("Monthly Income Distribution by Attrition")
plt.show()

# -----------------------------
# 4. Box Plot: Job Satisfaction vs Attrition
# -----------------------------
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Attrition', y='JobSatisfaction', palette='coolwarm')
plt.title("Job Satisfaction by Attrition")
plt.show()

# -----------------------------
# 5. Pie Chart: Education Field
# -----------------------------
plt.figure(figsize=(6, 6))
df['EducationField'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90,
                                             colors=plt.cm.Pastel1.colors)
plt.title("Distribution by Education Field")
plt.ylabel("")
plt.show()
```
