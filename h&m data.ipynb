```python

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
tfidf = TfidfVectorizer(max_features = 1000)
X_tfidf = tfidf.fit_transform(df['details_cleaned'])
pca = PCA(n_components = 2,random_state = 42)
X_pca = pca.fit_transform(X_tfidf.toarray())
plt.figure(figsize=(10,6))
plt.scatter(X_pca [:,0],X_pca[:,1],alpha = 0.5 , s = 20 , color = 'lightblue')
plt.title("PCA of IDF of details")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.tight_layout()
plt.show()
```

```python
df['details_cleaned'] = df['details'].fillna('').astype(str)
```

```python
#bigram analysis
vect_ngram = CountVectorizer(ngram_range = (2,2) , max_features = 20, stop_words='english')
X_ngrams = vect_ngram.fit_transform(df['details_cleaned'])
ngrams = vect_ngram.get_feature_names_out()
Count = X_ngrams.toarray().sum(axis=0)
plt.figure(figsize=(10, 5))
sns.barplot(x=Count, y=ngrams, palette='coolwarm')
plt.title('Top 20 Bigrams in details')
plt.xlabel('Frequency')
plt.ylabel('Bigram')
plt.tight_layout()
plt.show()

```

```python
kmeans = KMeans(n_clusters = 5, random_state = 42)
clusters = kmeans.fit_predict(X_tfidf)
df['cluster']= clusters
plt.figure(figsize=(10,6))
sns.scatterplot(x = X_pca[:,0],y=X_pca[:,1],hue = clusters , palette = 'Set2',s=30)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Kmeans clustering")
plt.tight_layout()
plt.show()

```
