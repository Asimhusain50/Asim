```python
from sklearn.preprocessing import StandardScaler
```

```python
features  = df[['Product Price','Total Purchase Amount']].dropna()
scalar = StandardScaler()
scaled_data = scalar.fit_transform(features)
scaled_df = pd.DataFrame(scaled_data , columns = ['Product Price' , 'Total Purchase Amount'])
```

```python
sns.scatterplot(data = scaled_df , x = 'Product Price' , y = 'Total Purchase Amount' , alpha = 0.5)
```

```python
from sklearn.decomposition import PCA
```

```python
features = df[['Product Price' , 'Total Purchase Amount']].dropna()
scalar = StandardScaler()
scalar_data = scalar.fit_transform(features)
pca = PCA(n_components = 2)
pca_result = pca.fit_transform(scalar_data)
pca_df = pd.DataFrame(pca_result , columns = ['PCA2' , 'PCA1'])
sns.scatterplot(pca_df , x = 'PCA1' , y = 'PCA2' , alpha = 0.5)
plt.title('PCA1 and PCA 2')
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid(True)
plt.show()
```
